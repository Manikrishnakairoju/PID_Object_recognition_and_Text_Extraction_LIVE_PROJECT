```python
import os
import cv2
import numpy as np
import pytesseract
from glob import glob

# 💡 Set your local Tesseract path
pytesseract.pytesseract.tesseract_cmd = r'C:\Program Files\Tesseract-OCR\tesseract.exe'

# 📁 Input/Output Directories
input_dir = r"C:\Users\aakas\OneDrive\Desktop\CAAP\CAAP\cropped_circles"
output_dir = r"C:\Users\aakas\OneDrive\Desktop\CAAP\CAAP\formatted_circles"
os.makedirs(output_dir, exist_ok=True)

# 📌 Set target size
TARGET_SIZE = 354

# 🛠️ Helper to format image to circular tag style
def format_circle_tag(image):
    # Resize to standard square size
    img_resized = cv2.resize(image, (TARGET_SIZE, TARGET_SIZE), interpolation=cv2.INTER_CUBIC)

    # Denoise and sharpen
    gray = cv2.cvtColor(img_resized, cv2.COLOR_BGR2GRAY)
    denoised = cv2.fastNlMeansDenoising(gray, None, h=30, templateWindowSize=7, searchWindowSize=21)

    # Sharpen filter
    kernel = np.array([[0, -1, 0],
                       [-1,  5,-1],
                       [0, -1, 0]])
    sharpened = cv2.filter2D(denoised, -1, kernel)

    # Adaptive threshold to make text pop
    final = cv2.adaptiveThreshold(sharpened, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C,
                                  cv2.THRESH_BINARY_INV, 15, 8)

    return final

# 🧠 OCR configuration (tweak if needed)
ocr_config = r'--oem 3 --psm 6 -c tessedit_char_whitelist=ABCDEFGHIJKLMNOPQRSTUVWXYZ0123456789-/ '

# 🔁 Process all cropped circle images
for img_path in glob(os.path.join(input_dir, "*.jpg")):
    filename = os.path.basename(img_path)
    image = cv2.imread(img_path)

    if image is None:
        print(f"❌ Failed to load: {filename}")
        continue

    formatted = format_circle_tag(image)

    # Save enhanced version
    out_path = os.path.join(output_dir, filename)
    cv2.imwrite(out_path, formatted)

    # OCR the final image
    text = pytesseract.image_to_string(formatted, config=ocr_config).strip()
    print(f"🔡 {filename} --> OCR: {text}")
```

```python

import os
import cv2
import numpy as np
import easyocr
from ultralytics import YOLO
```

```python
def enhance_image(image):
    # Upscale
    image = cv2.resize(image, (354, 354), interpolation=cv2.INTER_CUBIC)

    # Convert to grayscale
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # Equalize histogram
    equalized = cv2.equalizeHist(gray)

    # Apply adaptive thresholding
    processed = cv2.adaptiveThreshold(equalized, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C,
                                      cv2.THRESH_BINARY, 11, 2)
    return processed
```

```python
def ocr_and_save_circles(image_path, boxes, output_crop_dir, reader):
    img = cv2.imread(image_path)
    ocr_results = []

    for idx, box in enumerate(boxes):
        x1, y1, x2, y2 = map(int, box)
        cropped = img[y1:y2, x1:x2]

        # Enhance
        enhanced = enhance_image(cropped)

        # OCR
        result = reader.readtext(enhanced)
        text = result[0][1].strip() if result else 'no_text'

        # Sanitize filename
        safe_text = "".join(c for c in text if c.isalnum() or c in (' ', '_')).rstrip()
        crop_filename = f"{os.path.splitext(os.path.basename(image_path))[0]}_circle{idx}_{safe_text}.jpg"
        crop_path = os.path.join(output_crop_dir, crop_filename)

        # Save cropped image
        cv2.imwrite(crop_path, cropped)
        print(f"🟢 Saved: {crop_filename} | OCR: {text}")

        ocr_results.append((box, text))
    return ocr_results
```

```python
def run_pipeline(input_dir, model_path):
    model = YOLO(model_path)
    reader = easyocr.Reader(['en'], gpu=False)

    output_crop_dir = "cropped_circles"
    os.makedirs(output_crop_dir, exist_ok=True)

    for filename in os.listdir(input_dir):
        if not filename.lower().endswith(('.jpg', '.jpeg', '.png')):
            continue

        image_path = os.path.join(input_dir, filename)
        results = model.predict(source=image_path, conf=0.1, save=False, device='cpu')
        
        for result in results:
            boxes = result.boxes.xyxy.cpu().numpy()
            classes = result.boxes.cls.cpu().numpy()

            circle_boxes = [box for box, cls in zip(boxes, classes) if int(cls) == 0]
            if circle_boxes:
                print(f"\n🔍 Processing: {filename}")
                ocr_and_save_circles(image_path, circle_boxes, output_crop_dir, reader)
```

```python
# 📦 Required installations (run these in terminal if not already installed)
# pip install easyocr opencv-python-headless ultralytics

import os
import cv2
import easyocr
from ultralytics import YOLO

# Initialize YOLOv8 model and EasyOCR reader
model = YOLO("yolov8best.pt")
reader = easyocr.Reader(['en'], gpu=False)

# Input/output paths
input_dir = r"C:\Users\aakas\OneDrive\Desktop\CAAP\CAAP\input_images"
crop_dir = r"C:\Users\aakas\OneDrive\Desktop\CAAP\CAAP\cropped_tags"
os.makedirs(crop_dir, exist_ok=True)

# Function to preprocess image
def preprocess_image(img):
    img = cv2.resize(img, (0, 0), fx=5, fy=5, interpolation=cv2.INTER_CUBIC)
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    _, thresh = cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)
    return thresh

# Iterate through input images
for filename in os.listdir(input_dir):
    if not filename.lower().endswith(('.jpg', '.png', '.jpeg')):
        continue

    print(f"\n🔍 Processing: {filename}")
    img_path = os.path.join(input_dir, filename)
    image = cv2.imread(img_path)

    results = model(img_path)[0]

    for i, box in enumerate(results.boxes):
        x1, y1, x2, y2 = map(int, box.xyxy[0].tolist())

        pad = 10
        x1 = max(x1 - pad, 0)
        y1 = max(y1 - pad, 0)
        x2 = min(x2 + pad, image.shape[1])
        y2 = min(y2 + pad, image.shape[0])

        crop = image[y1:y2, x1:x2]
        processed = preprocess_image(crop)

        # OCR
        results = reader.readtext(processed, detail=0)
        text = " ".join(results).strip() if results else "no_text"

        # Save cropped image with result in name
        out_path = os.path.join(crop_dir, f"{filename.split('.')[0]}_circle{i}_{text[:10]}.jpg")
        cv2.imwrite(out_path, crop)
        print(f"🟢 Saved: {os.path.basename(out_path)} | OCR: {text}")

print("\n✅ Done. All circles processed and text extracted.")
```

```python
import cv2
import numpy as np
import os
from pathlib import Path
import easyocr

# === CONFIGURATION === #
INPUT_DIR = Path(r'C:\Users\aakas\OneDrive\Desktop\CAAP\CAAP\cropped_tags')      # Folder containing your cropped circles
OUTPUT_DIR = Path(r'C:\Users\aakas\OneDrive\Desktop\CAAP\CAAP\enhanced_circles')  # Folder to save enhanced images
RESIZE_DIM = (354, 354)                # Resize to match reference samples

# === SETUP === #
OUTPUT_DIR.mkdir(parents=True, exist_ok=True)
reader = easyocr.Reader(['en'], gpu=False)  # Use GPU=True if available

# === IMAGE ENHANCEMENT FUNCTION === #
def enhance_image(img):
 # Resize to 354x354
    img = cv2.resize(img, (454, 454), interpolation=cv2.INTER_CUBIC)

    # --- Step 1: Gamma Correction to fix overexposure ---
    gamma = 0.6  # Less than 1 darkens the image
    invGamma = 1.0 / gamma
    table = np.array([(i / 255.0) ** invGamma * 255 for i in np.arange(0, 256)]).astype("uint8")
    img = cv2.LUT(img, table)

    # --- Step 2: Convert to Grayscale ---
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

    # --- Step 3: Apply CLAHE (Contrast Limited Adaptive Histogram Equalization) ---
    clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(12, 12))
    eq = clahe.apply(gray)

    # --- Step 4: Sharpening Filter ---
    sharpen_kernel = np.array([[0, -1, 0],
                               [-1, 5, -1],
                               [0, -1, 0]])
    sharp = cv2.filter2D(eq, -1, sharpen_kernel)

    # --- Step 5: Bilateral Filter for Noise Reduction while Preserving Edges ---
    denoised = cv2.bilateralFilter(sharp, d=3, sigmaColor=65, sigmaSpace=65)

    # Convert back to BGR (if needed for OCR)
    final = cv2.cvtColor(denoised, cv2.COLOR_GRAY2BGR)

    return final


# === MAIN LOOP === #
for file in sorted(INPUT_DIR.glob("*.*")):
    if file.suffix.lower() not in ['.jpg', '.jpeg', '.png']:
        continue

    # Read and enhance the image
    img = cv2.imread(str(file))
    enhanced = enhance_image(img)

    # Perform OCR
    result = reader.readtext(enhanced, detail=0)
    text = "_".join(result).strip() if result else "no_text"

    # Save with OCR label
    output_path = OUTPUT_DIR / f"{file.stem}_{text}.png"
    cv2.imwrite(str(output_path), enhanced)

    print(f"🟢 Saved: {output_path.name} | OCR: {text}")
```
